```python
# Problem 1.1

import numpy as np
import seaborn as sns
from scipy import stats

x_bar = 215
mu_0 = 200
s = 25
n = 30
alpha = 0.05

t_stat = (x_bar-mu_0)/(s/(np.sqrt(n)))
p_value = 2 * (1- stats.t.cdf(abs(t_stat), df=n-1))

print('The test statistic is:', t_stat)
print('The p-value is:', p_value)
```
